```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from package import plot

def wrightFisher(N,p_0,generations):
    freqs = [p_0]
    for _ in range(generations):
        p = freqs[-1]
        sampled_alleles = np.random.binomial(2*N,p)
        freqs.append(sampled_alleles/(2*N))
    return freqs
```

```python
lst = wrightFisher(100,0.5,100)
```

```python
sns.lineplot(lst)
```

```python
import random

def generateRandomDNASeq(length):
    bases = list("ACGT")
    return "".join([random.choice(bases) for _ in range(length)])

def mutateDNASeq(sequence,mutation_rate):
    bases = list("ACGT")
    mutated_sequence = []
    for base in sequence:
        if random.random() < mutation_rate:
            new_base = random.choice([b for b in bases if b != base])
            mutated_sequence.append(new_base)
        else:
            mutated_sequence.append(base)
    return "".join(mutated_sequence)
```
